```python
company=['google', 'amazon', 'msft', 'fb']
revenue=[90, 123, 89, 27]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
years = [2001, 2005, 2010, 2015, 2020, 2025]
pandas = [2000, 1000, 500, 600, 1200, 3000]
young = [100, 59, 300, 200, 800, 1700]



plt.bar(years, pandas, width = 2, label = 'old')
plt.bar(years, young, width = 1.4, label='young' )
plt.ylabel('Number of pandas population')
plt.xlabel('Year')
plt.title('Trends in Pandas population')
plt.legend()
```

```python
plt.barh(years, pandas, label = 'old')
plt.barh(years, young,  label='young' )
plt.ylabel('Number of pandas population')
plt.xlabel('Year')
plt.title('Trends in Pandas population')
plt.legend()
```
